```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np
from skimage.feature import greycomatrix,greycoprops
import cv2
```

```python
image_directory="C:/Users/LENOVO/Desktop/Python/github/wood.jpg";
image=cv2.imread(image_directory,0)
imshow(image,cmap='gray')
```

```python
g = greycomatrix(image, [1], [0,45,90,135],levels=256,normed=True, symmetric=False)
correlation = greycoprops(g, 'correlation')
correlation=np.mean(correlation)
contrast=greycoprops(g, 'contrast')
contrast=np.mean(contrast)
homogeneity=greycoprops(g, 'homogeneity')
homogeneity=np.mean(homogeneity)
dissimilarity=greycoprops(g, 'dissimilarity')
dissimilarity=np.mean(dissimilarity)
ASM=greycoprops(g, 'ASM')
ASM=np.mean(ASM)
g_correlation = float("{:.3f}".format(correlation))
g_contrast = float("{:.3f}".format(contrast))
g_homogeneity = float("{:.3f}".format(homogeneity))
g_dissimilarity = float("{:.3f}".format(dissimilarity))
g_ASM = float("{:.3f}".format(ASM))

fig, ax =plt.subplots()
plt.subplots_adjust(left=0.7, bottom=0.5)
data=[[g_correlation],
      [g_contrast],
      [g_homogeneity],
     [g_dissimilarity],
     [g_ASM]]
column_labels=["Values"]
ax.axis('off')
ax.table(cellText=data,colLabels=column_labels,rowLabels=["Correlation","Contrast","Homogeneity","Dissimilarity","ASM"],loc="left")
plt.show()
```
